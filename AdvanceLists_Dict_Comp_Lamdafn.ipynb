# slicing
* list[start:end:step]
* start----inclusive
* stop-----exclusive
* step-----jump size
* Negative step → reverse

```python
lst = list(range(1, 21))
sli_ce = lst[5:16] # from index 5 to 15
extracted = lst[::3]# every 3rd element
reverse = lst[::-1]# reversed list
last_five = lst[-5:]# lastfive
remove = lst[1:-1]# remove 1 and last numbers
print("Indexes 5th to 15:",sli_ce)
print("Every 3rd element:", extracted)
print("Reversed:", reverse)
print("Last five elements:", last_five)
print("Excluded 1st and last elements:", remove)
```

With slicing:
list[start:end] = new_values
We can:
* Replace elements
* Insert elements
* Delete elements
* Even change list size

```python
#Replace middle 3 elements with [100, 200, 300]
data = [10, 20, 30, 40, 50, 60, 70]
data[2:5] = [100,200,300]
print("Replaced data:",data)
```

```python
#Insert [999, 888] at index 3 using slicing
data1 = [10, 20, 30, 40, 50, 60, 70]
data1[3:3] = [999,888] #data1[3:3] means “replace nothing at index 3” → so it inserts.
print("Inserted data:",data1)
```

```python
#Delete elements from index 2 to 4 using slicing
data2 = data1 = [10, 20, 30, 40, 50, 60, 70]
data2[2:5] = []
print("Deleted list:",data2)
```

# Packing and Unpacking
* (*) in unpacking → collects multiple values
* You can use only one starred variable
* Order matters

```python
#Unpack into a, b, *middle, e
numbers=[1,2,3,4,5]
a,b,*middle,e = numbers
print("a:",a)
print("b:",b)
print("Middle:",middle)
print("e:",e)
```

```python
#Print only the middle elements
numbers=[1,2,3,4,5]
a,b,*middle,e = numbers
print(middle)
```

```python
%%time
#Swap first and last using unpacking
#way1
numbers[0],numbers[-1]=numbers[-1],numbers[0] # pythonic way
print(numbers)
```

```python
%%time
#way2
#various ways of swapping 
numbers1=[1,2,3,4,5]
first, *mid, last = numbers1
numbers1 = [last, *mid, first]
print(numbers1)
```

```python
%%time
#way3
numbers=[1,2,3,4,5]
numbers = [e,b,*middle,a]
print(numbers)
```

# List Comprehensions
* General pattern [expression for item in iterable if condition]
* Break it mentally as:
* What to generate?
* From where?
* Under what condition?

```python
#Create a list of:
#Squares of even numbers from 1–50
squares_even = [x**2 for x in range(1,51) if x % 2 == 0] 
print(squares_even)
```

```python
#Cubes of numbers divisible by 5
cubes = [x**3 for x in range(1,51) if x % 5 == 0]
print(cubes)
```

```python
#Numbers divisible by 3 and 7 between 1–100
divisibilty = [x for x in range(1,101) if x % 3 == 0 and x % 7 == 0]
print(divisibilty)
```

```python
#Squares of odd numbers divisible by 3 from 1–100
squares_odd = [x**2 for x in range(1,101) if x%2 != 0 and x%3==0]
print(squares_odd)
```

list comprehension with Enumerate ------(index, value)

from this ["apple", "banana", "cherry"]
to this :
(0, "apple")
(1, "banana")
(2, "cherry")

```python
#Basic example
words = ["apple", "banana", "cherry"]
indexed_words = [(i,word) for i, word in enumerate(words)]
print(indexed_words)
```

```python
#starting from index 1
indexed_words = [(i,word) for i, word in enumerate(words, start = 1)]
print(indexed_words)
```

```python
#Get words with even index
words = ["apple", "banana", "cherry"]
even_indexed_words = [word for i, word in enumerate(words) if i % 2 ==0]
print(even_indexed_words)
```

List comprehension with zip

[expression for var1, var2 in zip(list1, list2) if condition]


```python
#Combine two lists
names2 = ["Ram", "Shyam", "Geeta"]
marks = [85, 90, 78]
combined = [(name,mark) for name,mark in zip (names2,marks)]
print(combined)
```

```python
#Students who scored > 80
top_students = [name for name,mark in zip(names2,marks) if mark > 80]
print(top_students)
```

```python
#Advanced: zip() + enumerate() Together
names = ["Ram", "Shyam", "Geeta"]
marks = [85, 90, 78]
result = [(i,name,mark) for i, (name,mark) in enumerate(zip(names,marks),start = 1)]
print(result)
```

```python
#formatted results
Formatted = [f"{i}.{name} scored {mark}"
             for i, (name,mark) in enumerate(zip(names,marks),start = 1)]
print(Formatted)
```

```python
#types of formatting ['Math -> 88', 'Physics -> 92', 'Chemistry -> 79']
subjects = ["Math", "Physics", "Chemistry"]
scores = [88, 92, 79]
formatted = [f"{subject} -> {score}"
             for (subject,score) in zip(subjects,scores)]
print(formatted)
```

```python
#Create list of word lengths
words = ["apple", "banana", "cherry", "date"]
word_length = [len(word) for word in words]
print(word_length)
```

```python
#Create list of uppercase words
uppercase = [word.upper() for word in words]
print(uppercase)
```

```python
#Create list of words starting with 'b'
start_b = [word for word in words if word.startswith('b')]
print(start_b)
```

```python
#create list of words length greater than 5
words = ["apple", "banana", "cherry", "kiwi", "watermelon", "grape"]
gret_5 = [len(word) for word in words if len(word) > 5]
print(gret_5)
```

# Nested Lists

```python
#Extract second row
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
second_row = matrix[1]
print(second_row)
```

```python
#Extract third column
third_column = [row[2] for row in matrix]
print(third_column)
```

```python
#Flatten the matrix into single list
flattened = [element for row in matrix for element in row]
print(flattened)
```

```python
#Transpose the matrix using list comprehension
transpose = [[row[i] for row in matrix] for i in range(len(matrix[0]))]
print(transpose)
```
