# slicing
* list[start:end:step]
* start----inclusive
* stop-----exclusive
* step-----jump size
* Negative step → reverse

```python
lst = list(range(1, 21))
sli_ce = lst[5:16] # from index 5 to 15
extracted = lst[::3]# every 3rd element
reverse = lst[::-1]# reversed list
last_five = lst[-5:]# lastfive
remove = lst[1:-1]# remove 1 and last numbers
print("Indexes 5th to 15:",sli_ce)
print("Every 3rd element:", extracted)
print("Reversed:", reverse)
print("Last five elements:", last_five)
print("Excluded 1st and last elements:", remove)
```

# List Comprehensions
* General pattern [expression for item in iterable if condition]
* Break it mentally as:
* What to generate?
* From where?
* Under what condition?

```python
#Create a list of:
#Squares of even numbers from 1–50
squares_even = [x**2 for x in range(1,51) if x % 2 == 0] 
print(squares_even)
```

```python
#Cubes of numbers divisible by 5
cubes = [x**3 for x in range(1,51) if x % 5 == 0]
print(cubes)
```

```python
#Numbers divisible by 3 and 7 between 1–100
divisibilty = [x for x in range(1,101) if x % 3 == 0 and x % 7 == 0]
print(divisibilty)
```

```python
#Squares of odd numbers divisible by 3 from 1–100
squares_odd = [x**2 for x in range(1,101) if x%2 != 0 and x%3==0]
print(squares_odd)
```

list comprehension with Enumerate ------(index, value)

from this ["apple", "banana", "cherry"]
to this :
(0, "apple")
(1, "banana")
(2, "cherry")

```python
#Basic example
words = ["apple", "banana", "cherry"]
indexed_words = [(i,word) for i, word in enumerate(words)]
print(indexed_words)
```

```python
#starting from index 1
indexed_words = [(i,word) for i, word in enumerate(words, start = 1)]
print(indexed_words)
```

```python
#Get words with even index
words = ["apple", "banana", "cherry"]
even_indexed_words = [word for i, word in enumerate(words) if i % 2 ==0]
print(even_indexed_words)
```

List comprehension with zip

[expression for var1, var2 in zip(list1, list2) if condition]


```python
#Combine two lists
names2 = ["Ram", "Shyam", "Geeta"]
marks = [85, 90, 78]
combined = [(name,mark) for name,mark in zip (names2,marks)]
print(combined)
```

```python
#Students who scored > 80
top_students = [name for name,mark in zip(names2,marks) if mark > 80]
print(top_students)
```

```python
#Advanced: zip() + enumerate() Together
names = ["Ram", "Shyam", "Geeta"]
marks = [85, 90, 78]
result = [(i,name,mark) for i, (name,mark) in enumerate(zip(names,marks),start = 1)]
print(result)
```

```python
#formatted results
Formatted = [f"{i}.{name} scored {mark}"
             for i, (name,mark) in enumerate(zip(names,marks),start = 1)]
print(Formatted)
```

# Nested Lists

```python
#Extract second row
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
second_row = matrix[1]
print(second_row)
```

```python
#Extract third column
third_column = [row[2] for row in matrix]
print(third_column)
```

```python
#Flatten the matrix into single list
flattened = [element for row in matrix for element in row]
print(flattened)
```

```python
#Transpose the matrix using list comprehension
transpose = [[row[i] for row in matrix] for i in range(len(matrix[0]))]
print(transpose)
```

# Sorting & Custom Sorting

* sorted(list, key= something)

* Means:

   👉 “Sort based on something.”

* Examples:
  
  key=len → sort by length

  
  key=lambda x: x[-1] → sort by last character

```python
#Sort by length
words = ["apple", "banana", "kiwi", "cherry"]
sort_by_length = sorted(words, key=len) #key=len → sort using word length.
print("Sorted order: ",sort_by_length)
```

```python
#Sort in reverse alphabetical order
words = ["apple", "banana", "kiwi", "cherry"]
sort_by_reverse = sorted(words, reverse = True) #reverse=True → descending order.
print("sortted by reverse: ",sort_by_reverse)
```

```python
#Sort by last character
sort_by_last_char = sorted(words, key=lambda word: word[-1]) #
print(sort_by_last_char)
```

# Filtering

```python
#Filter numbers divisible by 7
nums_fil = list(range(1,51))
nums_sev = [x for x in nums_fil if x % 7 == 0]
print(nums_sev)
```

```python
#Filter prime numbers
primes =[]

for num in nums_fil:
    if num > 1:
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                break
        else:
            primes.append(num)
print(primes)
```

# Real Logical Problems

```python
#Remove duplicates from: Without using set().
lst = [1,2,2,3,4,4,5,6,6,7]

unique = []
for num in lst:
    if num not in unique:
        unique.append(num)
print(unique)
```

```python
#Find second largest number in: Without using sort().
lst4 = [10, 20, 4, 45, 99]

largest = second = float('-inf') #float('-inf') means negative infinity.

for num in lst4:
    if num > largest:
        second = largest
        largest = num
    elif num > second and num != largest:
        second = num
print(second)
```

# Dictionaries

🔹 1️⃣A dictionary stores data in----> key : value
   * Keys must be:

     Unique
     
     Immutable (string, int, tuple)
🔹 2️⃣ Accessing Values
       Normal way:student["name"]
       ⚠ Problem: If key doesn’t exist → error
       Safer way:
          student.get("name")
          student.get("grade", "Not Available")


       If key not found → returns None or default value.

🔹 3️⃣ Adding & Updating
       Same syntax for both:

       student["branch"] = "CSE"   # Add
       student["marks"] = 90       # Update


      If key exists → update
      If key doesn’t exist → add
🔹 4️⃣ Removing Items
      Delete specific key
         del student["age"]

      pop() → removes & returns value
         student.pop("marks")

      Remove last inserted item
         student.popitem()
🔹 5️⃣ Important Dictionary Methods
      ✅ keys()

      Returns all keys

        student.keys()

      ✅ values()

      Returns all values

       student.values()

      ✅ items()

      Returns key-value pairs

       student.items()
      ✅ update()

      Merge another dictionary

        student.update({"city": "Hyderabad"})

      ✅ clear()

        Removes everything

        student.clear()
🔹 6️⃣ Looping Through Dictionary
       Only keys:
         for key in student:
             print(key)

       Only values:
         for value in student.values():
             print(value)

       Both key and value:
         for key, value in student.items():
             print(key, value)
🔹 7️⃣ Dictionary Comprehension

       Syntax:

       {key_expression: value_expression for item in iterable}


       Example:

       nums = [1,2,3]
       square_dict = {x: x*x for x in nums}
🔥 Most Important Concepts to Remember

* Keys are unique

* Access using [] or get()

* items() for looping

* get() prevents errors

* Dictionary comprehension saves time

👍 dictionaries are extremely important because:

JSON data → dictionaries

API responses → dictionaries

Feature storage → dictionaries

Model parameters → dictionaries

Frequency counting → dictionaries

```python
#Basic Creation & Access (Foundation)
student = {
    "name": "Ram",
    "age": 20,
    "marks": 85
}
#Print only age
print("Age: ",student["age"])

#Add new key "branch": "CSE"
student["branch"] = "CSE"
print(student)
#Update marks to 90
student["marks"]=90
print(student)
#Remove age
del student["age"]
print(student)
#Print all keys
print(student.keys())
#Print all values
print(student.values())
```

```python
#Frequency Counter (VERY Important for ML)
#words = text.split()----splits text
text = "machine learning is amazing and machine learning is powerful"
words = text.split()
print(words)
freq = {}
for word in words:
    freq[word] = freq.get(word, 0)+1
print(freq)

#freq.get(word, 0)
#→ If word exists, get its value
#→ If not, return 0
#Then add 1
```

```python
from collections import Counter #----important Logic
freq = Counter(words)
print(freq)
```

```python
#Find Key with Maximum Value (Common Interview Question)
scores = {
    "Ram": 85,
    "Shyam": 90,
    "Geeta": 78
}
#Find student with highest score
top_student = max(scores, key = scores.get) #----->max(dictionary, key=dictionary.get)
print(top_student)
print(scores[top_student])

#Find student with lowest score
low_student = min(scores, key = scores.get) #----->min(dictionary, key=dictionary.get)
print(low_student)
print(scores[low_student])
```

```python
scores = { #----cleaner way/pythonic way
    "Ram": 85,
    "Shyam": 90,
    "Geeta": 78
}
name, score = max(scores.items(), key=lambda x: x[1])
print(name,"-->", score)
#x[0] → key
#x[1] → value
```

```python
#Nested Dictionary (ML Style Data)
#dictionary[key1][key2]
#First level → outer
#Second level → inner

students = {
    "Ram": {"score": 85, "attendance": 90},
    "Shyam": {"score": 90, "attendance": 95},
    "Geeta": {"score": 78, "attendance": 80}
}
#Print Shyam’s attendance
print("Shyam's attendance: ",students["Shyam"]["attendance"])

#Increase every student's score by 5
for student in students:
    students[student]["score"] += 5
print(students)

#Print students who scored > 85
for name,details in students.items():
    if details ["score"] > 85:
        print(name)
```

```python
# Dictionary Comprehension (Feature Engineering Style)
#-----{key : value for item in iterable}

nums = [1, 2, 3, 4, 5]
#key = number
#value = "even" or "odd"
result_EO = {num: ("even" if num % 2 == 0 else "odd") for num in nums}
print(result_EO)
```

```python
#Invert a Dictionary (Advanced Thinking) ----Very important grouping logic (used in ML grouping tasks)
data = {
    "a": 1,
    "b": 2,
    "c": 1,
    "d": 2
}
result = {}
for key, value in data.items():
    if value not in result:
        result[value] = []
    result[value].append(key)
print(result)
```

```python
#using get()
result = {}

for key, value in data.items():
    result[value] = result.get(value, []) + [key]

print(result)
```

```python
#Sort Dictionary by Value ------> sorted(dictionary.items(), key=lambda x: x[1])

#---Ranking systems
#---ML prediction sorting
#---Frequency analysis
#---Leaderboards
marks = {
    "Ram": 85,
    "Shyam": 90,
    "Geeta": 78
}
#Sort dictionary by marks (ascending)
sort_asc = dict(sorted(marks.items(), key= lambda x: x[1])) #x[0] → key (Ram)  x[1] → value (85)------>We tell Python: 👉 “Sort based on value.”
print(sort_asc)
#Sort dictionary by marks (descending)
sorted_desc = dict(sorted(marks.items(), key=lambda x: x[1], reverse=True))
print(sorted_desc)
```
