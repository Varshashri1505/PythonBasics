Matplotlib:



*   Useful for making plots


```python
# import matplotlib library
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
# adding title x-axis nd y-axis labels
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Sine value')
plt.title('Sine wave')
plt.show()
```

```python
#parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'g+')
plt.show()
```

```python
plt.plot(x,y,'r.')
plt.show()
```

```python
plt.plot(x,y,'px')
plt.show()
```
