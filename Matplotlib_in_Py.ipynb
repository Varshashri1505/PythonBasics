Matplotlib:

```python
# import matplotlib library
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
# adding title x-axis nd y-axis labels
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Sine value')
plt.title('Sine wave')
plt.show()
```

```python
#parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'g+')
plt.show()
```

```python
plt.plot(x,y,'r.')
plt.show()
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.bar(languages,people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.pie(people, labels = languages, autopct='%1.1f%%')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='r')
ax.scatter(x,z,color='g')
plt.show()
```

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Greens')
plt.show()
```

```python
#standarad syntaxes before every visualtizations
import matplotlib.pyplot as plt
import numpy as np
```

Important plot structure
* plt.plot(x, y)-----Graph needs:X-axis values, Y-axis values
* plt.xlabel("X label")----name for axis
* plt.ylabel("Y label")----name for axis
* plt.title("Title")-------title name
* plt.show()-----give output cauz before it stores in memory
* plt.figure(figsize=(8,5))------size of chart
* plt.grid()--------adds reference lines
* plt.legend()--------when multiple lines exists, Legend tells which line represents what.

**Line plot**

Used when:

* Data changes over time

* Continuous values

  
Used for:

 * Trends

 * Time data

 * Growth analysis

```python
#Basic level
x = [1,2,3,4]
y = [10,20,25,30]

plt.plot(x,y)
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales Growth")
plt.show()
```

```python
#adding stye
x = [1,2,3,4]
y = [10,20,25,30]
plt.plot(x,y,color = 'green',marker = 'o',linestyle = '--')
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales Growth")
plt.show()
```

```python
tests = [1,2,3,4,5]
marks = [60,65,70,80,90]

plt.plot(tests,marks)
plt.xlabel("test number")
plt.ylabel("Marks")
plt.title("Student Progress")
plt.show()
```

```python
#Multiple Lines---------Compare multiple datasets.
math = [70,75,80,85]
science = [60,70,75,82]

plt.plot(math,label="Maths",marker='o',color='purple')
plt.plot(science,label="Science",marker='o',color='orange')
plt.legend()
plt.title("Subject Comparision")
plt.show()
```
